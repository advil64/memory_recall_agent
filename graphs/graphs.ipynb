```python
import pandas as pd
```

```python
invis_builds = pd.read_csv('/content/invisible_buildings.csv')
spot_diff = pd.read_csv('/content/spot_diff.csv')

invis_builds.head()
```

```python
import plotly.express as px

entropy_loss_invis_builds = px.line(invis_builds[invis_builds['time/total_timesteps'] < 800000], x='time/total_timesteps', y='train/entropy_loss', title='Invisible Goal with Buildings Training')
value_loss_invis_builds = px.line(invis_builds[invis_builds['time/total_timesteps'] < 800000], x='time/total_timesteps', y='train/value_loss', title='Invisible Goal with Buildings Training')
policy_loss_invis_builds = px.line(invis_builds[invis_builds['time/total_timesteps'] < 800000], x='time/total_timesteps', y='train/policy_loss', title='Invisible Goal with Buildings Training')
rew_mean_invis_builds = px.line(invis_builds[invis_builds['time/total_timesteps'] < 800000], x='time/total_timesteps', y='rollout/ep_rew_mean', title='Invisible Goal with Buildings Training')
```
